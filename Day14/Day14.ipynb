```python
import numpy as np
import copy
```

```python
rules = {}
with open('input14.txt', 'r') as inf:
    template = [c for c in inf.readline().strip()]
    inf.readline()
    for line in inf.readlines():
        rules[(line[0], line[1])] = line[6]

for step in range(0, 10):
    new_temp = []
    for i in range(0, len(template)-1):
        pattern = (template[i], template[i+1])
        if pattern in rules:
            new_temp += [pattern[0], rules[pattern]]
        else:
            new_temp += [pattern[0]]
    
    new_temp += [pattern[-1]]

    template = new_temp
    print('Step', step+1, len(template))
    
template = np.array(template)
elements = np.unique(np.array(template))

max_elem = ''
max_elem_c = 0

min_elem = ''
min_elem_c = 1e9

for elem in elements:
    n_elem = len(np.where(template == elem)[0])
    if n_elem > max_elem_c:
        max_elem = elem
        max_elem_c = n_elem
    if n_elem < min_elem_c:
        min_elem = elem
        min_elem_c = n_elem
        
print(max_elem, max_elem_c)
print(min_elem, min_elem_c)

print('Result', max_elem_c - min_elem_c)
```

```python
# Part 2
rules = {}

c_map = {}
pattern_map = {}

with open('input14.txt', 'r') as inf:
    template = [c for c in inf.readline().strip()]
    inf.readline()
    for line in inf.readlines():
        rules[(line[0], line[1])] = line[6]

for c in rules.values():
    c_map[c] = 0

for c in template:
    c_map[c] += 1
    
for r in rules:
    pattern_map[r] = 0

for i in range(0, len(template)-1):
    pattern = (template[i], template[i+1])
    if pattern in rules:
        pattern_map[pattern] += 1
    
for step in range(0, 40):
    temp_c_map = copy.copy(c_map)
    temp_pattern_map = copy.copy(pattern_map)
    for p in rules:
        # p is a pattern, e.g. (C, H)
        n_patterns = temp_pattern_map[p]
        if n_patterns > 0:
            # This pattern is in the template
            # Add the added characters
            c_map[rules[p]] += n_patterns
            # Add new resulting patterns
            if (p[0], rules[p]) in rules:
                # We create a new pattern
                pattern_map[(p[0], rules[p])] += n_patterns
            if (rules[p], p[1]) in rules:
                # We create a new pattern
                pattern_map[(rules[p], p[1])] += n_patterns
            pattern_map[p] -= n_patterns

    print('Step', step+1, sum(c_map.values()))
    
max_elem = ''
max_elem_c = 0

min_elem = ''
min_elem_c = 1e20

for elem, n_elem in c_map.items():
    if n_elem > max_elem_c:
        max_elem = elem
        max_elem_c = n_elem
    if n_elem < min_elem_c:
        min_elem = elem
        min_elem_c = n_elem
        
print(max_elem, max_elem_c)
print(min_elem, min_elem_c)

print('Result', max_elem_c - min_elem_c)
```
