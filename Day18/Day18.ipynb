```python
import numpy as np
import copy
```

```python
def flatten(data, curr_idx, idx, struct):
    for i in (0, 1):
        next_idx = copy.copy(curr_idx)
        next_idx.append(i)
        if isinstance(data[i], list):
            flatten(data[i], next_idx, idx, struct)
        else:            
            idx.append(next_idx)
            struct.append(data[i])
            
def get_val(data, idx):
    if len(idx) > 1:
        return get_val(data[idx[0]], idx[1:])
    else:
        return data[idx[0]]


def set_val(data, idx, val):
    if len(idx) > 1:
        set_val(data[idx[0]], idx[1:], val)
    else:
        data[idx[0]] = val


def add_val(data, idx, val):
    if len(idx) > 1:
        add_val(data[idx[0]], idx[1:], val)
    else:
        data[idx[0]] += val
        
def get_magnitude(data):
    lval = data[0]
    rval = data[1]
    if isinstance(data[0], list):
        lval = get_magnitude(data[0])
    if isinstance(data[1], list):
        rval = get_magnitude(data[1])
        
    return 3 * lval + 2 * rval

def reduce(data):
    reduced = False
        
    while not reduced:
            
        flat_structure = []
        flat_idx = []
            
        flatten(data, [], flat_idx, flat_structure)

        # print('IDX', flat_idx, 'S', flat_structure, 'D', data)
        # Find explosions
        reduced = True
        for i, idx in enumerate(flat_idx):
                
            if len(idx) > 4:
                # print('Data', data)
                # print('I', i, 'IDX', idx)
                # Explode number
                left = get_val(data, idx)
                right = get_val(data, flat_idx[i+1])
                    
                if i>0:
                    add_val(data, flat_idx[i-1], left)
                if i+1 < len(flat_idx)-1:
                    add_val(data, flat_idx[i+2], right)
                set_val(data, idx[:-1], 0)
                    
                # print('Now', data)
                reduced = False
                break
                    
        if not reduced:
            continue

        # Find splits
        for idx in flat_idx:
            # print('Spl', data, idx)
            val = get_val(data, idx)
            if val > 9:
                # print('Splitting', data, val)
                left = int(np.floor(val / 2))
                right = int(np.ceil(val / 2))
                set_val(data, idx, [left, right])
                    
                # print('Result', data)
                reduced = False
                break
```

```python
numbers = []
with open('input18d.txt', 'r') as inf:
    for line in inf.readlines():
        numbers.append(eval(line, {}))
        
    for n in numbers[1:]:
        data = [copy.deepcopy(numbers[0]), copy.deepcopy(n)]
        
        reduce(data)
                
print('Final', data)    

# Calc magnitude

print('Magnitude', get_magnitude(data))

print('Part 2')

max_mag = 0

for i in range(0, len(numbers)):
    for j in range(i+1, len(numbers)):
        data = [copy.deepcopy(numbers[i]), copy.deepcopy(numbers[j])]

        reduce(data)
        mag = get_magnitude(data)

        max_mag = max(max_mag, get_magnitude(data))
        
        data = [copy.deepcopy(numbers[j]), copy.deepcopy(numbers[i])]

        reduce(data)
        mag = get_magnitude(data)

        max_mag = max(max_mag, get_magnitude(data))

print('Max Mag', max_mag)
```
